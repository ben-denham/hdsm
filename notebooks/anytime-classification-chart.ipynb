```clojure
(import '[java.awt Color]
        '[com.twosigma.beakerx.chart.legend LegendPosition LegendLayout]
        '[com.twosigma.beakerx.chart.xychart Plot CombinedPlot]
        '[com.twosigma.beakerx.chart.xychart.plotitem Points ShapeType])

(def colors [(Color/decode "#ff96a8") (Color/decode "#f55774") (Color/decode "#d12249") (Color/decode "#1b2126")])
(def outline-colors [(Color/black) (Color/black) (Color/black) (Color/black)])
(def shapes [ShapeType/CIRCLE ShapeType/TRIANGLE ShapeType/SQUARE ShapeType/DIAMOND])
(def sizes [10 13 10 10])

(let [acc-plot (doto (Plot.)
                   (.setLegendPosition LegendPosition/TOP)
                   (.setLegendLayout LegendLayout/HORIZONTAL)
                   (.setYLabel "Accuracy")
                   (.setXBound [0.5 3.5])
                   (.setYBound [0.65 0.95]))
      mrtcp-plot (doto (Plot.)
                     (.setShowLegend false)
                     (.setYLabel "× Order 0 MTCP")
                     (.setXBound [0.5 3.5])
                     (.setYBound [15 55]))
      mrtbc-plot (doto (Plot.)
                     (.setShowLegend false)
                     (.setYLabel "× Order 0 MTBC")
                     (.setXBound [0.5 3.5])
                     (.setYBound [10 20]))
      plot (doto (CombinedPlot.)
               (.add acc-plot)
               (.add mrtcp-plot)
               (.add mrtbc-plot)
               (.setInitWidth 450)
               (.setInitHeight 750)
               (.setXLabel "Order"))]
    (doseq [[[label order-summaries] outline-color color shape size]
            (map vector results outline-colors colors shapes sizes)]
        (let [order-zero-summary (first (vals order-summaries))
              order-summaries (rest order-summaries)]
            (.add acc-plot
                  (doto (Points.)
                      (.setX (keys order-summaries))
                      (.setY (map :accuracy (vals order-summaries)))
                      (.setShape shape)
                      (.setColor color)
                      (.setOutlineColor outline-color)
                      (.setSize size)
                      #_(.setFill false)
                      (.setDisplayName label)))
            (.add mrtcp-plot
                  (doto (Points.)
                      (.setX (keys order-summaries))
                      (.setY (map #(/ (:mrtcp %) (:mrtcp order-zero-summary)) (vals order-summaries)))
                      (.setShape shape)
                      (.setColor color)
                      (.setOutlineColor outline-color)
                      (.setSize size)
                      #_(.setFill false)
                      (.setDisplayName label)))
            (.add mrtbc-plot
                  (doto (Points.)
                      (.setX (keys order-summaries))
                      (.setY (map #(/ (:mrtbc %) (:mrtbc order-zero-summary)) (vals order-summaries)))
                      (.setShape shape)
                      (.setColor color)
                      (.setOutlineColor outline-color)
                      (.setSize size)
                      #_(.setFill false)
                      (.setDisplayName label)))))
    plot)
```
